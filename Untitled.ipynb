# Pipeline

```python
df = pd.DataFrame()

# Create a column
df['name'] = ['Kumar', 'Anil', 'Tom']
df['gender'] = ['Male', 'Male', 'Female']
df['maths'] = [31, 32, 19]
df['physics'] = [31, 32, 19]
df['chemestry'] = [31, 32, 19]
```

```python
df
```

```python
def avg_by_group(dataframe, col):
    # groups the data by a column and returns the mean of subject per group
    return dataframe.groupby(col).mean()
def uppercase_column_name(dataframe):
    # Capitalizes all the column headers
    dataframe.columns = dataframe.columns.str.upper()
    # And returns them
    return dataframe

(df.pipe(avg_by_group, col='gender')
   # then applies the uppercase column name function
   .pipe(uppercase_column_name)
)
```
